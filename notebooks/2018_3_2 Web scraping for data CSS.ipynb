https://sites.google.com/a/chromium.org/chromedriver/downloads

Move the chromedriver to a path , e.g. "chromedriverpath"

```python
from selenium import webdriver

chromedriverpath = '/Users/dan/Desktop/Intern_2018/chromedriver'
browser = webdriver.Chrome(chromedriverpath)

css_cat_path2 = 'https://www.chowsangsang.com/en/product/Jewellery-ByDesign-Flower?page=4'

url_2 = css_cat_path2
browser.get(url_2) #navigate to the page
innerHTML_2 = browser.execute_script("return document.body.innerHTML") #returns the inner HTML as a string
```

## 2. Parse using BeautifulSoup 

```python
import bs4
from bs4 import BeautifulSoup as soup
import pandas as pd

page_soup_2 = soup(innerHTML_2, "html.parser")

#get the name of the products 
name_raw = page_soup_2.findAll("div",{"class":"comp-pdt-blk-name"})
name_list = []
for item in name_raw:
    name_list.append(item.text)

#get the descriptions of the products
material_raw = page_soup_2.findAll("div",{"class":"comp-pdt-blk-material"})
material_list = []
for item in material_raw:
    material_list.append(item.text)

#get the price of the products
price_raw = page_soup_2.findAll("span",{"class":"price-new"})
price_list = []
for item in price_raw:
    price_list.append(item.text)

#get the images of the products 
image_raw = page_soup_2.findAll("div",{"class":"comp-pdt-blk-img-container"})
image_list = []
for item in image_raw:
    image_list.append(item.img['src'])  


d = {"Product_title": name_list, "Product_description":material_list,\
    "Price":price_list, "Images":image_list}
css_df = pd.DataFrame(data=d)
css_df = css_df.reindex_axis(['Product_title','Product_description','Price','Images'], axis=1)
css_df
```
