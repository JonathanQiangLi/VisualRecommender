## 1. Use Selenium to scrape javascript rendered page

http://stanford.edu/~mgorkove/cgi-bin/rpython_tutorials/Scraping_a_Webpage_Rendered_by_Javascript_Using_Python.php

Download chromedriver from: 
https://sites.google.com/a/chromium.org/chromedriver/downloads

Move the chromedriver to a path , e.g. "chromedriverpath"

```python
from selenium import webdriver

chromedriverpath = '/Users/dan/Desktop/Intern_2018/chromedriver'
cartier_cat_path = 'http://www.cartier.com/en-us/collections/jewelry/collections/juste-un-clou.viewall.html'

browser = webdriver.Chrome(chromedriverpath) #replace with .Firefox(), or with the browser of your choice
url = cartier_cat_path
browser.get(url) #navigate to the page

innerHTML = browser.execute_script("return document.body.innerHTML") #returns the inner HTML as a string
```

```python
len(innerHTML)
```

## 2. Parse using BeautifulSoup

```python
import bs4
from bs4 import BeautifulSoup as soup

page_soup = soup(innerHTML, "html.parser")
wrapper = page_soup.findAll("div",{"class":"col-sm-3 col-xs-6 grid-item-parent"}) 

print(len(wrapper))
```

### 1). Get the images address

```python
#http://www.cartier.com
image_list = []
for item in wrapper:
    image = item.find("div",{"class":"image-block"})
    image_str = str(image).split("data-original-url=")[1]
    image_new =image_str.split('"')[1]
    image_list.append(str("http://www.cartier.com")+image_new)
   
```

### 2). Get the information including category, price, metal, collections, stone shape 

```python
price_list = []
description_list = []
for item in wrapper:
    item1 = item.find("div",{"class":"grid-item"})
    data_price = item1['data-price']
    price_list.append(data_price)
    
    item2 = item1['filter']    
    description_list.append(item2)
```

```python
import pandas as pd
import ast

desc_list_new = []
for item in description_list:
    description_dict = ast.literal_eval(item.replace("'",'').replace('[','').replace(']',''))
    description_dict.values()
    desc_list_new.append(list(description_dict.values()))
key = ast.literal_eval(description_list[4].replace("'",'').replace('[','').replace(']',''))
columns = key.keys()
desc_list_new

df_1 = pd.DataFrame(data=desc_list_new)
df_1.columns = columns
```

```python
d_2 = {"Price":price_list,"Image":image_list}
df_2 = pd.DataFrame(data=d_2)
```

```python
catier_df = pd.concat([df_1, df_2], axis=1)
catier_df.columns = ['Collections','Category','Metal','Stone shape','Price','Image']
catier_df
```
